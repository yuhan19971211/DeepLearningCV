```python
import numpy as np
import skimage.io as io
from skimage import morphology
from skimage import data
import matplotlib.pyplot as plt
img=data.checkerboard()
plt.imshow(img,plt.cm.gray)
```

```python
#膨胀
img_d=morphology.dilation(img,morphology.square(15))
plt.imshow(img_d,plt.cm.gray)
```

```python
#腐蚀
img_e=morphology.erosion(img,morphology.square(20))
plt.imshow(img_e,plt.cm.gray)
```

```python
#开运算
img_o=morphology.opening(img,morphology.square(25))
plt.imshow(img_o,plt.cm.gray)
```

```python
#闭运算
img_c=morphology.closing(img,morphology.square(25))
plt.imshow(img_c)
```

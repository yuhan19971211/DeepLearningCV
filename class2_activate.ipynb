```python
#转换灰白图片
img_gray=np.mean(img2,2)#取平均
plt.imshow(img_gray)
#调用函数
img_gray2=skimage.color.rgb2gray(img2)
plt.imshow(img_gray2)
```

```python
#转换为HSV
img_hsv=skimage.color.rgb2hsv(img2)
plt.imshow(img_hsv)
#转换回来
#hsv2rgb(img_hsv)
```

```python
#根据像素值的修改来进行相关的绘图
img2[50:150,100:150,:]=255
plt.imshow(img2)
```

```python
#调用skimage包绘制
from skimage import draw
h,w=draw.line(200,250,100,50)
img2[h,w,0]=255
img2[h,w,1]=0
img2[h,w,2]=0
plt.imshow(img2)
```

```python
#画圆
img2=io.imread('data/cat.jpg')
h,w=skimage.draw.circle(50,100,40)#最后一个是半径40
draw.set_color(img2,[h,w],[255,100,50])
plt.imshow(img2)
```

```python
#图像的平移旋转变换等
img2=io.imread('data/cat.jpg')
from skimage import transform#引入变换库
img_r=transform.resize(img2,output_shape=[300,200])#resize尺寸
plt.imshow(img_r)
```

```python
img2=io.imread('data/cat.jpg')
img_r=transform.rotate(img2,45,resize=True)#rotate 旋转角度，不加resize=ture,出去的部分就不会保留，true就可以通过padding保留
plt.imshow(img_r)
#transform 还有很多其他的指令
```

```python
#图像亮度变化
from skimage import exposure
img2=io.imread('data/cat.jpg')
img_g=exposure.adjust_gamma(img2,1)#数字越大越暗，越小越亮
plt.imshow(img_g)
```

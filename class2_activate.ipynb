```python
#
img2=io.imread('data/cat.jpg')
h,w=skimage.draw.circle(50,100,40)#40
draw.set_color(img2,[h,w],[255,100,50])
plt.imshow(img2)
```

```python
#
img2=io.imread('data/cat.jpg')
from skimage import transform#
img_r=transform.resize(img2,output_shape=[300,200])#resize
plt.imshow(img_r)
```

```python
img2=io.imread('data/cat.jpg')
img_r=transform.rotate(img2,45,resize=True)#rotate resize=ture,truepadding
plt.imshow(img_r)
#transform 
```
